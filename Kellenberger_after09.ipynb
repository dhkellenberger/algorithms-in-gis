```python
#(1) Import numpy as np, matplotlib.pyplot as plt, and pandas as pd. Set matplotlib to inline display.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#(2) Read the Fisher Iris Data in as a variable called df.

df = pd.read_csv('fisher_iris_data.csv')
```

```python
#(3) Display the first five lines of df using head.

df.head(5)
```

```python
#(4-7) - Plot a histogram of each of the four variables (leaf/sepal length/width).  
#Adjust the number of bins to produce a resolution you find instructive/visually satisfying.  
#Use xlabel and ylabel to label the axes.  Each plot should be its own codeblock.

#(8) Save each figure using savefig.
```

```python
plt.hist(df.petal_width, 20)
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.savefig('iris_petal_width_hist.png', dpi=300)
```

```python
plt.hist(df.petal_length, 20)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.savefig('iris_petal_length_hist.png', dpi=300)
```

```python
plt.hist(df.sepal_width, 20)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.savefig('iris_sepal_width_hist.png', dpi=300)
```

```python
plt.hist(df.sepal_length, 20)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.savefig('iris_sepal_length_hist.png', dpi=300)
```

```python
#(9) Using groupby, group the data by species

groups = df.groupby(by='species')

#(10) Using the groups, create a bar graph of the mean sepal length by species.  Label the axes.

means = groups['sepal_length'].mean()
plt.bar(means.index, means)
plt.xlabel('Species')
plt.ylabel('Mean of Sepal Length')
```

```python
#(11) Use subplots to make a 2x2 pane figure, with each pane a scatterplot of your choosing.  
#Make sure axes are labeled!

plt.subplot(2,2,1)
plt.plot(df.petal_width, df.petal_length, '.')
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')

plt.subplot(2,2,2)
plt.plot(df.sepal_width, df.sepal_length, '.')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')

plt.subplot(2,2,3)
plt.plot(df.petal_width, df.sepal_width, '.')
plt.xlabel('Petal Width')
plt.ylabel('Sepal Width')

plt.subplot(2,2,4)
plt.plot(df.petal_length, df.sepal_length, '.')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')

plt.subplots_adjust(wspace=.5, hspace=.5)
```

```python
#(12) Use tail to print out the last 10 rows of df.

df.tail(10)
```

```python
# (13) Use a combination of np.sum() and a selector (e.g., (df['x'] > 3)
# Count how many rows have a petal length greater than 40.

x = np.sum(df['petal_length'] > 40)
print(x)
```

```python
#(14) How many have a sepal width below the median sepal width?
x = np.sum(df['sepal_width'] < np.median(df['sepal_width']))
print(x)
```

```python
#(15) Use groups to report the mean values for each variable (by species).

groups = df.groupby(by='species')
means = groups.mean()
means
```

```python
#(16) Calculate a new variable based on existing variables, and create a histogram of its distribution.

df['sepal_thinness'] = df.sepal_length / df.sepal_width

_ = plt.hist(df.sepal_thinness, 20)
```

```python
#(17) Export the data with the new calculated variable as an Excel file. Open it in Excel, 
#and comment on its appearance. Look at the documentation of to_excel and figure out how to 
#make it so the index values (line numbers) don't get exported.

df.to_excel('after09.xlsx')
```

The data seems to have the species listed in order by petal length. Setosas have the smallest petal length, then versicolor, and finally virginicas. This seems to correlate to their sepal sizes and the sepal thickness as well. 

In order to export the data frame without the index values, there is an attribute of the "to_excel" function called "index" which takes a boolean value. If this is set to False, then the index #'s will not be exported to the table.

```python
url = 'https://raw.githubusercontent.com/thomaspingel/geodata/master/election/state_election_data_1976-2020.csv'
election_df = pd.read_csv(url)
election_df.head()
```

```python
#(18-22) Use subplots to make a 2x2 pane figure, with each pane a scatterplot of your choosing. 
#Make sure axes are labeled!

plt.subplot(2,2,1)
plt.plot(election_df.totalvotes_2020, election_df.gop_2020_votes, '.')
plt.xlabel('Total 2020 Votes')
plt.ylabel('GOP 2020 Votes')

plt.subplot(2,2,2)
plt.plot(election_df.totalvotes_2020, election_df.dem_2020_votes, '.')
plt.xlabel('Total 2020 Votes')
plt.ylabel('Dem 2020 Votes')

plt.subplot(2,2,3)
plt.plot(election_df.totalvotes_1976, election_df.gop_1976_votes, '.')
plt.xlabel('Total 1976 Votes')
plt.ylabel('GOP 1976 Votes')

plt.subplot(2,2,4)
plt.plot(election_df.totalvotes_1976, election_df.dem_1976_votes, '.')
plt.xlabel('Total 1976 Votes')
plt.ylabel('Dem 1976 Votes')

plt.subplots_adjust(wspace=.5, hspace=.5)
```
